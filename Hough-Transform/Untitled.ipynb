```python
#this was for the sobel filter but were using the built in canny one now.
dxkernel = np.matrix([[-1,-1,-1], [0,0,0], [1,1,1]])
dykernel = np.matrix([[-1,0, 1], [-1,0, 1], [-1,0, 1]])
print(img.shape)

def convolution(img, kernel):

    retVal = np.zeros((img.shape[0], img.shape[1]))
    
    
    for y in range(2, img.shape[0]-2):
        for x in range(2,img.shape[1] -2):
            
            retVal[y][x] = ((img[y-1][x-1] * kernel[0,0]) + (img[y-1][x] * kernel[0,1]) + (img[y-1][x+1] * kernel[0,2]) +
                            (img[y]  [x-1] * kernel[1,0]) + (img[y]  [x] * kernel[1,1]) + (img[y]  [x+1] * kernel[1,2]) +
                            (img[y+1][x-1] * kernel[2,0]) + (img[y+1][x] * kernel[2,1]) + (img[y+1][x+1] * kernel[2,2])) 
                      
    return retVal

dx = convolution(edges, dxkernel)
dy = convolution(edges, dykernel)
```

```python
def getDiagLength(img):
    width, height = img.shape
    return np.ceil(np.sqrt( (width*width) + (height*height)))
```

```python
def k_largest_index_argpartition(a, k):
    idx = np.argpartition(-a.ravel(),k)[:k]
    return np.column_stack(np.unravel_index(idx, a.shape))
```

```python
def hough(img):
    thetas = np.arange(-90,91,5)
    
    #diagonal length of the image
    diag = int (getDiagLength(img))
    rhos = np.linspace(-diag, diag, diag * 2 +1)
    num_thetas = len(thetas)
    accumulator = np.zeros((2 * diag+1, num_thetas))
    
    #stores the row col indexes of the edges in the image as these are all we need to iterate over.
    y_idxs, x_idxs = np.nonzero(img)

    #calc now to save time in the forloop
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    #for each edge in image
    for i in range(len(x_idxs)):
        #get coordinates of the edge
        x = x_idxs[i]
        y = y_idxs[i]
        
        #for each theta we generated above
        for t_idx in range(num_thetas):
            # Calculate rho. or p diag_len is added for a positive index

            rho = int (round(x * cos_t[t_idx] + y * sin_t[t_idx]) )
            
            accumulator[rho, t_idx] += 1

    return (accumulator, thetas, rhos)

#call the function and get the accumulator our new feature space , thetas to know the angle accuracy and rhos for the indexs
accumulator, thetas, rhos = hough(img)


```
